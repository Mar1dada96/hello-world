# Write a Pandas program to create and display a DataFrame from the following dictionary data which has index labels.

```python
import numpy as np
```

```python
import pandas as pd
```

```python
exam_data = {'name':['Anastasia','Dima','Katherin','James','Emily','Michael',
                    'Mathew','Laura','Kevin','Jonas'], 
             'score':[12.5,9,16.5,np.nan,9,20,14.5,np.nan,8,19],
             'attempts':[1,3,2,3,2,3,1,1,2,1],
             'qualify':['yes','no','yes','no','no','yes','yes','no','no','yes']}
df = pd.DataFrame(exam_data)
df
```

```python
df = df.set_index('name')
df
```

# Delete rows with Nan values

```python
df = df.dropna()
df
```

# Write a Pandas program to append a new row 'k' to data frame with these values.

```python
k = {'name':'Suresh','score':15.5,'attempts':1,'qualify':'yes'}
new_std = pd.DataFrame(k,index = ['Suresh'])
new_std.set_index('name')
```

# Write a Pandas program to delete the 'attempts' column from the DataFrame.

```python
full_lst = pd.concat([df,new_std])
full_lst.drop(columns=['name','attempts'])
```

# Add a new column "Success" : if the score is higher than 10 we will have 1 else we will have 0 

```python
full_lst['success'] = [int(True) if x>=10 else int(False) for x in full_lst['score']]
```

```python
full_lst = full_lst.drop(columns='name')
```
